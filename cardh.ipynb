Formatting the columns

```python
import re
import numpy as np
```

```python
import pandas as pd

# Simple function to convert price to float
def convert_price(price):
    if pd.isna(price):
        return None
    
    price = str(price).replace('₹', '').replace(',', '').strip()
    
    if 'Lakh' in price:
        return float(price.replace('Lakh', '').strip()) * 100000
    elif 'Crore' in price:
        return float(price.replace('Crore', '').strip()) * 10000000
    else:
        return float(price)

# Apply the function
car_details_6states['Price'] = car_details_6states['Price'].apply(convert_price)
```

```python
# Function to clean engine displacement
def clean_engine(cc):
    if pd.isna(cc):
        return np.nan
    return str(cc).replace('cc', '').strip()

# Apply the function
car_details_6states['Engine_Displacement(cc)'] = car_details_6states['Engine_Displacement(cc)'].apply(clean_engine)
```

```python
import re

# Function to clean mileage
def mileage_conv(x):
    if isinstance(x, str):
        match = re.search('kmpl$', x)
        if match:
            try:
                return float(x.rstrip('kmpl').strip())
            except ValueError:
                return None
    return x  # return original value if it's already float or doesn't match

    # Apply the function
car_details_6states['Mileage(kmpl)'] = car_details_6states['Mileage(kmpl)'].apply(mileage_conv)
```

```python
def convert_power_to_hp(x):
    if pd.isna(x):
        return np.nan

    x = str(x).lower().strip()

    # Extract the first number (handles decimals)
    match = re.findall(r'\d+\.?\d*', x)
    if not match:
        return np.nan

    value = float(match[0])

    # Convert based on unit
    if 'ps' in x:
        return round(value * 0.98632, 2)
    elif 'kw' in x:
        return round(value * 1.34102, 2)
    else:
        # Assume bhp/hp or no unit is already in hp
        return round(value, 2)
    
# Apply the function
car_details_6states['Max Power(bhp)'] = car_details_6states['Max Power(bhp)'].apply(convert_power_to_hp)    
```

```python
def clean_kms(kms):
    if pd.isna(kms):
        return np.nan
    return int(str(kms).replace(',', '').strip())

car_details_6states['Kms_Driven'] = car_details_6states['Kms_Driven'].apply(clean_kms)
```
